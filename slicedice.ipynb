```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_json('train.json')
```

```python
print "There are %i cuisine classes" % (len(df['cuisine'].unique()))
print "The class labels are: ", df['cuisine'].unique()
```

```python
def all_words(df):
    matrix = []
    for i in xrange(len(df)):
        row = df.iloc[i]
        cuisine = row['cuisine']
        _id = row['id']
        ingredients = row['ingredients']
        for ingredient in ingredients:
            matrix.append([_id, cuisine, ingredient])
    newdf = pd.DataFrame(matrix, columns=['id', 'cuisine', 'ingredient'])
    return newdf
```

```python
dfi = all_words(df)
```

```python
print "Out of %i ingredient entries, there are %i unique ingredients" % (len(dfi), len(dfi['ingredient'].unique()))
```

```python
def topwords(N, cuisine, df):
    blah = df.groupby(['cuisine', 'ingredient'], as_index=False).count().values
    blah = pd.DataFrame(blah, columns=['cuisine', 'ingredient', 'count'])
    return blah[blah['cuisine']==cuisine].sort(columns='count', ascending=False).iloc[0:N]
```

```python
topwords(20, 'moroccan', dfi)
```
